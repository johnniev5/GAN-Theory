默认GAN使用的是JS距离来衡量G和D之间的距离，而fGAN则可以选择其它的距离来衡量...

在使用JS距离的时候，如果两个分布一致，则其JS距离为0，而有一点不同，距离就大于0，所以说无论你怎么train，只要两个分布不一致，它们的JS距离就不为0，也就无法收敛，换句话说，只要没有一致，train的效果都是一样的...



如果$f(x) = x \log x$，$f^*(t)=exp(t-1)$








使用不同的距离可以解决使用JS带来的Mode Collapse问题，就是生成器的分布分不开，而判别器的分得很开的问题... 小一点的问题就是Mode dropping...
