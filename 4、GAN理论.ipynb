我们在使用GAN时，假定它是满足一个分布的，在这个分布中生成的对象都是可靠的，而不在这个空间中的那就是不好的。

找一个生成模型去最大化似然，其实是等价于最小化生成模型和实际分布之间的KL距离。

如果在知道生成模型和真实数据的分布，是可以使用，比如最小化KL距离等求出来的，但是现在都不知道，该如何做呢？

就是通过神经网络来训练，使用GAN，也就说判别器代表着真实样本的分布，而生成器代表着要生成的，目的就是最大化样本的真实分布，最小化生成的概率，这样就能够实现生成的样本尽量近似于真实样本... 这就是精妙所在... 如下最大化(G,D)之间的距离，其实就是最大化JSD距离...
